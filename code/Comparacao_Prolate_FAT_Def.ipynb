```python
import numpy as np
from matplotlib import pyplot as plt
from fatiando.gravmag import ellipsoid_prolate
from fatiando import mesher, gridder, utils

%matplotlib inline
```

```python
import ellipsoid_prolate_matrix as me2
```

```python
# Geographic coordinates grid
xmin = -1000.
xmax = 1000.
ymin = -900.
ymax = 900.
Nx = 200
Ny = 200

# Create a regular grid at 0m height
shape = (Nx, Ny)
area = [xmin, xmax, ymin, ymax]
Xp, Yp, Zp = gridder.regular(area, shape, z=0.)

# Set the inclination and declination of the regional field
inten, inc, dec  = 60000., 90., 0.

#xc position x; yc position y; and zc depth
xc = 0.
yc = 0.
zc = 100.

#Remanent vector
intensity_rem = 10000.
incli_rem = 90.
decli_rem = 0.

#susceptibility directions
intensidadeA = 0.2
intensidadeB = 0.1
intensidadeC = 0.05
k = [intensidadeA,intensidadeB,intensidadeC,0.0,90.0,90.0]

azimuth = 0.
delta = 0.

daxis = 20.
aref = 50.
```

# Implementation  -  Clark's Ellipsoid Triaxial Function

```python
# ellipsoid axis
a = aref + daxis
b = aref

Bx_T, By_T, Bz_T,Tf_T = me2.ellipsoid (Xp,Yp,Zp,xc,yc,zc,a,b,azimuth,delta,decli_rem,incli_rem,intensity_rem,dec,inc,inten,k)
```

# Implementation  -  Clark's Ellipsoid Triaxial FAT

```python
model = [mesher.EllipsoidProlate(xc,yc,zc,a,b,azimuth,delta, 
                                 {'remanence': [intensity_rem, incli_rem, decli_rem],
                                  'k': [intensidadeA, intensidadeB, intensidadeC, 0., 90., 90.]})]

Bx = ellipsoid_prolate.bx_c(Xp,Yp,Zp,inten,inc,dec,model)
By = ellipsoid_prolate.by_c(Xp,Yp,Zp,inten,inc,dec,model)
Bz = ellipsoid_prolate.bz_c(Xp,Yp,Zp,inten,inc,dec,model)
Tf = ellipsoid_prolate.tf_c(Xp,Yp,Zp,inten,inc,dec,model)
```

## Results

### Plot Bx

```python
rangesBx = np.max(np.abs([np.max(Bx), np.min(Bx),
                          np.max(Bx_T), np.min(Bx_T)]))

plt.figure(figsize=(15,8))

plt.subplot(1,2,1)
plt.title('Ellipsoid Prolate Class', y=1.08)
plt.axis('scaled')
plt.tricontourf(Yp,Xp,Bx,15,
            vmin = -rangesBx, vmax = rangesBx,
            cmap=plt.cm.RdBu_r)
cb = plt.colorbar(shrink=0.7)
cb.set_label('$Bx$ ( $nT$ )', rotation=90, fontsize=14)
plt.xlim(ymin,ymax)
plt.ylim(xmin,xmax)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Horizontal coordinate y (m)', fontsize=14)
plt.ylabel('Horizontal coordinate x (m)', fontsize=14)

plt.subplot(1,2,2)
plt.title('Ellipsoid Prolate Function', y=1.08)
plt.axis('scaled')
plt.tricontourf(Yp,Xp,Bx_T,15,
            vmin = -rangesBx, vmax = rangesBx,
            cmap=plt.cm.RdBu_r)
cb = plt.colorbar(shrink=0.7)
cb.set_label('$Bx$ ( $nT$ )', rotation=90, fontsize=14)
plt.xlim(ymin,ymax)
plt.ylim(xmin,xmax)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Horizontal coordinate y (m)', fontsize=14)
plt.ylabel('Horizontal coordinate x (m)', fontsize=14)

plt.tight_layout()
```
